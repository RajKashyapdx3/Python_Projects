```python
df[df['Category']=='Breakfast'].sort_values(by='Calories',ascending=False)
```

```python
df_break=df[df['Category']=='Breakfast'].sort_values(by='Calories',ascending=False)
df_break
```

```python
df_break[['Item','Calories']].sort_values(by='Calories',ascending=False).head(10)
```

```python
df_break['Item'].value_counts().index
```

```python
plt.figure(figsize=(18,16))
df_break['Item'].value_counts().index
ax=sns.barplot(x='Item',y='Calories',data=df_break)
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()),(p.get_x()+0.1,p.get_height()+2))
plt.xticks(rotation=60);
```
